#Running Meshroom
1. If you use Drive and already have an existing directory for your output, delete the first line of code.
2. If you don't have an output directory in Drive, keep the code: it will create an output directory in the Colab archive space.
3. Use the File tab to navigate to your input directory, right click, copy path and paste it in the code below (in place of `./input`).
4. Repeat step 3 for the output directory, pasting the path in place of `./output`.
5. Run the code

```
!mkdir output
!./meshroom/Meshroom-2021.1.0-av2.4.0-centos7-cuda10.2/meshroom_batch -i ./input -o ./output
```
